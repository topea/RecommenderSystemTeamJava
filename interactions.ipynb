```python
import pandas as pd
import numpy as np
import math
```

```python
user = 'https://raw.githubusercontent.com/Nedu/RecommenderSystemTeamJava/master/data/users.csv'
userData = pd.read_csv(user)
userData
```

```python
notifications = 'https://raw.githubusercontent.com/Nedu/RecommenderSystemTeamJava/master/data/notifications.csv'
notificationsData = pd.read_csv(notifications)
notificationsData
```

```python
posts = 'https://raw.githubusercontent.com/Nedu/RecommenderSystemTeamJava/master/data/posts.csv'
postsData = pd.read_csv(posts)
postsData
```

```python
userData['InteractedPosts'] = ''
tempNotifications = notificationsData

tempNotifications['post_id'].fillna(0.0, inplace=True)
tempNotifications = tempNotifications.astype({"post_id": int})

notificationsFiltered = tempNotifications[tempNotifications.post_id != 0]
notificationsFiltered
# notificationsFiltered[notificationsFiltered == userData.id].post_id
# userData['InteractedPosts'] = notificationsFiltered[notificationsFiltered == userData.id].post_id
# userData
```

```python
# notificationData['Ratings'] = notificationData['action'].apply(lambda x: action_rating[x])

# notificationData.head()

# from notifications.csv there are 5 reactions
event_type_strength = {
    'Followed': 1,
    'Like': 3,
    'Love': 4,
    'Commented': 5,
    'Replied': 4
}
notificationsData['eventStrength'] = notificationsData['action'].apply(lambda x: event_type_strength[x])
users_interactions_count_df = notificationsData.groupby(['user_id', 'post_id']).size().groupby('user_id').size()
x = (notificationsData[['user_id','post_id','eventStrength']].sort_values(by='user_id', axis= 0, ascending= True))

users_with_enough_interactions_df = users_interactions_count_df[users_interactions_count_df >= 2].reset_index()[['user_id']]

interactions_from_selected_users_df = notificationsData.merge(users_with_enough_interactions_df, how='right',
                                                            left_on='user_id', right_on='user_id')


def smooth_user_preference(x):
    return math.log(1 + x, 2)


interactions_full_df = interactions_from_selected_users_df \
    .groupby(['user_id', 'post_id'])['eventStrength'].sum() \
    .apply(smooth_user_preference).reset_index()
interactions_full_df['eventStrength'] = interactions_full_df['eventStrength'].astype(int)
# interactions_full_df['post_id'] = interactions_full_df['post_id'].astype(int)
# print(interactions_full_df)

item_popularity_df = interactions_full_df.groupby('post_id')['eventStrength'].sum().sort_values(ascending=True).reset_index()
# print(item_popularity_df.head(10))

item_popularity_df
x['post_id'] = x['post_id'].astype(int)
x = x[x['post_id'] != 0]
x
```

```python
x.to_csv('ratings.csv', index=None)
```
